```python
from konlpy.tag import Twitter
from gensim.models import Word2Vec
```

```python
twitter = Twitter()

def twitter_tokenizer(statement):
    tokens = twitter.pos(statement, norm=True, stem=True)
    return tokens
```

```python
file_name = './data/terms_pdf.txt'

lines = ''
tokenized_contents = []
with open(file_name, 'r') as file:
    tags = ['Foreign', 'Alpha'] # 'Punctuation', 
    
    for line in file.readlines():
        line = line[:-1]
        lines += line
        
    tokens = twitter_tokenizer(lines)
    tokenized_contents.append([token[0]+'//'+token[1] for token in tokens if token[1] not in tags])
```

```python
tokenized_contents
```

```python
final_datas = []
final_data = []
for content in tokenized_contents[0]:
    final_data.append(content)
    if './/Punctuation' == content:        
        final_datas.append(final_data)
        final_data = []
        
final_datas
```

```python
embedding_model = Word2Vec(final_datas, size=50, window = 2, min_count=50, workers=4, iter=1000, sg=1)
```

```python
embedding_model.wv.vocab
```

```python
embedding_model.wv.similar_by_vector("연체//Noun")
```

```python
embedding_model.save('./word2vec_model/embedding_50d_final.model')
```
